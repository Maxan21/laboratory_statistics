```python
sorted_data1 = sorted(data1)

efr = []
n = len(sorted_data1)
for i in range(n):
    efr.append((i + 1) / n)


plt.plot(sorted_data1, efr, marker='o')


plt.xlabel('Значения')
plt.ylabel('Процент')
plt.title('Эмпирическая функция распределения data1')
```

```python
import statistics
import scipy


mean = statistics.mean(data1)
mean_np = np.mean(data1)
print("Выборочное среднее data1:", mean, mean_np)

variance = statistics.variance(data1)
variance_np = np.var(data1, ddof=1)
print("Выборочная дисперсия data1:", variance, variance_np)

median = statistics.median(data1)
median_np = np.median(data1)
print("Медиана data1:", median, median_np)

mode = statistics.mode(data1)
print("Мода data1:", mode)

skewness = scipy.stats.skew(data1)
print("Коэффициент асимметрии data1:", skewness)

kurtosis = scipy.stats.kurtosis(data1)
print("Эксцесс data1:", kurtosis)
```

По полученному значению коэффициента ассиметрии можно сказать, что распределение не симметрично (т.к. оно не равно нулю). Значение коэффициента больше нуля, значит мы имеем дело с положительной ассиметрией.
Коэффициент эксцесса отрицательный, значит, распределение является плосковершинным, то есть значения величин распределены далеко от средней вершины.

```python
import statistics
import scipy

mean = statistics.mean(data2)
mean_np = np.mean(data2)
print("Выборочное среднее data2:", mean, mean_np)

variance = statistics.variance(data2)
variance_np = np.var(data2, ddof=1)
print("Выборочная дисперсия data2:", variance, variance_np)

median = statistics.median(data2)
median_np = np.median(data2)
print("Медиана data2:", median, median_np)

mode = statistics.mode(data2)
print("Мода data2:", mode)

skewness = scipy.stats.skew(data2)
print("Коэффициент асимметрии data2:", skewness)

kurtosis = scipy.stats.kurtosis(data2)
print("Эксцесс data2:", kurtosis)
```

По полученным значениям коэффициента ассиметрии и эксцесса понятно, что это не нормальное распределение, положительная ассиметрия. Коэффициент эксцесса положительный, что означает, что распределение обладает ярко выраженным пиком.
